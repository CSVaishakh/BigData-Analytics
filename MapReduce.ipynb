```python
from functools import reduce
from collections import defaultdict
import string
```

```python
s = "Artificial Intelligence is a field of Computer Science, Machine Leanring and Data Science are subsets of Artificial Intelligence"

clean_s = s.translate(str.maketrans('','',string.punctuation)).lower()
words = clean_s.split()

print(f"original sentance : {s}")
print(f"words in the sentance : {words}")
```

```python
def mapper(word):
    return (word, 1)

mapped_output = list(map(mapper, words))

print("1. Map Phase")
print(mapped_output)
```

```python
grouped_data = defaultdict(list)
for word,count in mapped_output:
    grouped_data[word].append(count)

print("2. Shuffle Phase")
for word, counts_list in grouped_data.items():
    print(f"{word} : {counts_list}")
```

```python
def reducer(accumulated_sum, next_value):
    return accumulated_sum + next_value

final_output = {}
for word,counts in grouped_data.items():
    total = reduce(reducer, counts)
    final_output[word] = total

print("3. Reduction(or Reduce) Phase")
for word, count in final_output.items():
    print(f"'{word}' : {count}")
```
