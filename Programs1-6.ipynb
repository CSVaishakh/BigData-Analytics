```python
def minMax():
    numbers = [10,15,1,5,7,25,30]
    for i in range(len(numbers)):
        for j in range(len(numbers)-1):
            if numbers[i] > numbers[j]:
                numbers[i],numbers[j] = numbers[j],numbers[i]
    print(f"Maximum Value = {numbers[0]}")
    print(f"Minimum Value = {numbers[len(numbers)-1]}")

minMax()
```

```python
def multiplicationTable():
    num = int(input("Enter the number whose multiplicant is to be displayed : "))
    r = int(input("Enter the no multipicants needed : "))
    for i in range(r+1):
        print(f"{num} * {i} = {num*i}")

multiplicationTable()
```

```python
def pascalsTriangle():
    n = 5
    for i in range(n):
        print(" "*(n-i),end = " ")
        num = 1
        for j in range(i+1):
            print(num, end = " ")
            num = num*(i-j)//(j+1)
        print()
pascalsTriangle()
```

```python
def insertElement():
    ln = int(input("Enter the size of the list : "))
    numbers = []
    for i in range(ln):
        num = int(input("Enter the number to be inserted : "))
        numbers.append(num)
    print(numbers)
insertElement()
```

```python
def factorial(): 
    n = int(input("Enter the number whose factorial is to be found : "))
    fact = 1
    for i in range(1,n+1):
        fact *= i
    print(f"Factorial of {n} is {fact}")
factorial()
```

```python
def palindrome():
    str = input("Enter the word to be checked for palindrome : ")
    rev = ""
    for ch in str:
        rev = ch+rev
    if str == rev: 
        print(f"The word {str} is a palindrome")
    else:
        print(f"The word {str} is not a palindrome")
palindrome()
```
