```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

variance = df.var()
covariance = df.cov()
correlation = df.corr()

print("Variance of Attributes:\n")
print(variance, "\n")

print("Covariance Matrix:\n")
print(covariance, "\n")

print("Correlation Matrix:\n")
print(correlation)
```
